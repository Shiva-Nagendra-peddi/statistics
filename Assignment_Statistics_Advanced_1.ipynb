**Question 1: What is a random variable in probability theory?**
Ans:
In probability theory, a random variable is a function that assigns a numerical value to each outcome of a random experiment.

Think of it like this:
You perform an experiment with uncertain results (rolling a die, flipping a coin, measuring rainfall tomorrow, etc.).
The experiment‚Äôs sample space lists all possible outcomes.
The random variable takes each outcome and turns it into a number you can work with mathematically.

**Question 7: What is the significance of confidence intervals in statistical analysis?**
Ans:
A confidence interval (CI) is a range of values, derived from sample data, that is likely to contain the true population parameter (such as the mean or proportion) with a certain level of confidence.

Instead of giving a single estimate (like the sample mean), we provide a range that accounts for sampling variability.
Example:
Sample mean = 50
95% confidence interval = (47.5, 52.5)
This means: We are 95% confident the true population mean lies between 47.5 and 52.5.

Why Confidence Intervals Are Important
Measure of precision ‚Äì Narrow intervals indicate more precise estimates.
Account for uncertainty ‚Äì Shows how much variability is expected from sampling.
More informative than point estimates ‚Äì Provides a range, not just one number.
Basis for decision-making ‚Äì If a CI for a mean difference does not include 0, it suggests a statistically significant difference.
Used in hypothesis testing ‚Äì CIs can serve as an alternative way to interpret test results.

**Question 8: What is the concept of expected value in a probability distribution?**
Ans:
The expected value (also called the mean or expectation) of a probability distribution is the long-run average outcome you would expect if you repeated a random experiment many times.
It represents the center or balance point of the distribution.

Why It‚Äôs Important
Decision-making under uncertainty: Choose actions with the highest expected payoff.
Finance: Expected return on investments.
Gambling & games: Fairness of bets.
Statistics: Basis for variance, standard deviation, and many other measures.

**Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.**
Ans:
Program Below

```python
#Q9 Prpgram Ans
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate 1000 random numbers from N(50, 5^2)
mean = 50
std_dev = 5
size = 1000

data = np.random.normal(mean, std_dev, size)

# Step 2: Compute sample mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # ddof=1 for sample std deviation

# Step 3: Print results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")

# Step 4: Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Histogram of Normal Distribution (mean=50, std=5)")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.axvline(sample_mean, color='red', linestyle='dashed', linewidth=2, label=f"Mean = {sample_mean:.2f}")
plt.legend()
plt.show()
```

**Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend. Daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,235, 260, 245, 250, 225, 270, 265, 255, 250, 260]**
‚óè Explain how you would apply the Central Limit Theorem to estimate the average saleswith a 95% confidence interval.
‚óè Write the Python code to compute the mean sales and its confidence interval.
Ans:
Applying the Central Limit Theorem (CLT)
We have daily sales data for a sample (here, 20 days are shown, but real case = 2 years).
We want to estimate the population mean and a 95% confidence interval (CI).

Also Program below

```python
#Q10 Program

import numpy as np
import scipy.stats as stats

# Given daily sales data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Sample statistics
mean_sales = np.mean(daily_sales)
std_sales = np.std(daily_sales, ddof=1)  # sample standard deviation
n = len(daily_sales)

# 95% confidence interval using t-distribution (safer for small samples)
confidence_level = 0.95
alpha = 1 - confidence_level
t_critical = stats.t.ppf(1 - alpha/2, df=n-1)  # t-critical value

# Margin of error
margin_of_error = t_critical * (std_sales / np.sqrt(n))

# Confidence interval
ci_lower = mean_sales - margin_of_error
ci_upper = mean_sales + margin_of_error

# Output
print(f"Mean Sales: {mean_sales:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```
